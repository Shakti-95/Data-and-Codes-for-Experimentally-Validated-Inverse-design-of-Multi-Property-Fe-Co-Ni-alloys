```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_excel('Multi-property models results_MAE.xlsx', index_col=0)
df1 = pd.read_excel('Multi-property models results_RMSE.xlsx', index_col=0)
df2 = pd.read_excel('Multi-property models results_R2.xlsx', index_col=0)
df3 = pd.read_excel('Multi-property models results Wen features_MAE.xlsx', index_col=0)
df4 = pd.read_excel('Multi-property models results Wen features_RMSE.xlsx', index_col=0)
df5 = pd.read_excel('Multi-property models results Wen features_R2.xlsx', index_col=0)
```

```python
mae_df = pd.concat([df.iloc[1,:],df.iloc[3,:],df3.iloc[1,:],df3.iloc[3,:]],axis=1)
rmse_df = pd.concat([df1.iloc[1,:],df1.iloc[3,:],df4.iloc[1,:],df4.iloc[3,:]],axis=1)
r2_df = pd.concat([df2.iloc[1,:],df2.iloc[3,:],df5.iloc[1,:],df5.iloc[3,:]],axis=1)
```

```python
xtl = [r'$M_s$', r'$log_{10}H_c$', r'$log_{10}T_c$', r'$log_{10}\rho$', r'$log_{10}\delta$', r'$log_{10}H_V$',
       r'$log_{10}Cost$']
```

```python
figure, ax = plt.subplots(1, 3, figsize=(39, 15))

ax[0] = mae_df.plot(kind="bar", ax=ax[0], colormap='Dark2')
ax[0].tick_params(length=10, width=1.5)
ax[0].set_ylabel(r'$MAE$', fontsize = 40, labelpad=25)
ax[0].set_xlabel('(g)', fontsize = 40, labelpad=25)
ax[0].set_xticklabels(xtl, rotation=70)
ax[0].legend(fontsize="25")
for label in (ax[0].get_xticklabels() + ax[0].get_yticklabels()):
    label.set_fontsize(32)

ax[1] = rmse_df.plot(kind="bar", ax=ax[1], colormap='Dark2')
ax[1].tick_params(length=10, width=1.5)
ax[1].set_ylim([0.0,0.7])
ax[1].set_ylabel(r'$RMSE$', fontsize = 40, labelpad=25)
ax[1].set_xlabel('(h)', fontsize = 40, labelpad=25)
ax[1].set_xticklabels(xtl, rotation=70)
ax[1].legend(fontsize="25")
for label in (ax[1].get_xticklabels() + ax[1].get_yticklabels()):
    label.set_fontsize(32)

ax[2] = r2_df.plot(kind="bar", ax=ax[2], colormap='Dark2')
ax[2].tick_params(length=10, width=1.5)
ax[2].set_ylim([0.0,1.3])
ax[2].set_ylabel(r'$R^2$', fontsize = 40, labelpad=25)
ax[2].set_xlabel('(i)', fontsize = 40, labelpad=25)
ax[2].set_xticklabels(xtl, rotation=70)
ax[2].legend(fontsize="25")
for label in (ax[2].get_xticklabels() + ax[2].get_yticklabels()):
    label.set_fontsize(32)

plt.tight_layout(pad=5)
plt.show()
```
