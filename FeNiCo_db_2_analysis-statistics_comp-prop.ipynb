```python
#Import libraries required for database analysis
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# Import database
db = pd.read_csv('FeNiCo_db_2.csv')
db
```

```python
db.info()
```

```python
# Creating a dataframe of composition only to analyze the distribution of elements
cmp = db.iloc[:,:15]
cmp
```

```python
cmp.describe()
```

```python
# Replacing 0 with NaN to get the proper distribution of elements
cmp1 = cmp.replace(0, np.nan)
cmp1
```

```python
# Creating a dataframe of properties only to analyze their distribution
prop = pd.concat([db.iloc[:,29:32], db.iloc[:,33:]], axis=1)
prop
```

```python
#Converting Tc values to Kelvin scale
prop.Tc = prop.Tc+273
prop
```

```python
prop.describe()
```

```python
# Visualizing the distribution of elements
comp_hist, axes = plt.subplots(5, 3, figsize=(24,28))

countX = 0
countY = 0
for var in cmp1:
    sb.histplot(x = cmp1[var], color = 'g', bins = 15, ax = axes[countX,countY], stat='count')
    axes[countX,countY].set_title('Distribution of '+var, fontsize=25)
    axes[countX,countY].set_xlabel('Atomic Percentage (at.%)', fontsize=25, labelpad=6)
    axes[countX,countY].set_ylabel('No. of Data Points', fontsize=25, labelpad=6)
    for label in (axes[countX,countY].get_xticklabels() + axes[countX,countY].get_yticklabels()):
        label.set_fontsize(20)
    countY += 1
    if countY == 3:
        countY = 0
        countX += 1
        
# set the spacing between subplots
plt.tight_layout()
plt.subplots_adjust(wspace=0.25,hspace=0.35)
plt.show()
```

```python
# Saving the plot of distribution of elements
comp_hist.savefig('Composition Histogram.png', dpi=300)
```

```python
# Visualizing the distribution of properties
prop_hist, axes = plt.subplots(3, 3, figsize=(30,20))

units = [r'$T$', r'$A/m$', r'$K$', r'$\mu\OmegaÂ·cm$', r'$MPa$', r'$MPa$', r'%', r'$HV$',
         r'$USD/kg$']
title = [r'$M_s$', r'$H_c$', r'$T_c$', r'$\rho$', r'$\sigma$', r'$\sigma_y$', r'$\delta$', r'$H_V$', r'$Cost$']
countX = 0
countY = 0
i=0
for var in prop:
    sb. histplot(x = prop[var], kde = True, color = 'r', ax = axes[countX,countY])
    axes[countX,countY].set_title('Distribution of '+title[i], fontsize=32, pad=12)
    axes[countX,countY].set_xlabel(units[i],fontsize=32, labelpad=6)
    axes[countX,countY].set_ylabel('No. of Data Points', fontsize=32, labelpad=6)
    for label in (axes[countX,countY].get_xticklabels() + axes[countX,countY].get_yticklabels()):
        label.set_fontsize(25)
    i += 1
    countY += 1
    if countY == 3:
        countY = 0
        countX += 1

# set the spacing between subplots
plt.tight_layout()
plt.subplots_adjust(wspace=0.24,hspace=0.37)        
plt.show()
```

```python
# Analyzing the skewness in distribution of each property
prop.skew()
```

```python
# Saving the plot of distribution of properties
prop_hist.savefig('Property Histogram.png', dpi=300)
```

```python
# Except Ms, converting all other properties into log scale to reduce skewness and to make different propery values comparable to each other
prop1 = pd.concat([prop.Ms,round(np.log10(prop.iloc[:,1:]),2)],axis=1)
prop1.rename(columns={'Hc':'logHc','Tc':'logTc','ER':'logER','TS':'logTS','YS':'logYS','Elong':'logEl','HV':'logHV','Cost':'logCost'},inplace=True)
prop1
```

```python
# Visualizing the distribution of modified properties
prop1_hist, axes = plt.subplots(3, 3, figsize=(30,20))

units = [r'$T$', r'$log_{10}(A/m)$', r'$log_{10}(K)$', r'$log_{10}(\mu\OmegaÂ·cm)$', r'$log_{10}(MPa)$', r'$log_{10}(MPa)$', r'$log_{10}(percent)$', r'$log_{10}(HV)$',
         r'$log_{10}(USD/kg)$']
title = [r'$M_s$', r'$log_{10}H_c$', r'$log_{10}T_c$', r'$log_{10}\rho$', r'$log_{10}\sigma$', r'$log_{10}\sigma_y$', r'$log_{10}\delta$', r'$log_{10}H_V$', r'$log_{10}Cost$']
countX = 0
countY = 0
i=0
for var in prop1:
    sb. histplot(x = prop1[var], kde = True, color = 'r', ax = axes[countX,countY])
    axes[countX,countY].set_title('Distribution of '+title[i], fontsize=32, pad=12)
    axes[countX,countY].set_xlabel(units[i],fontsize=32, labelpad=6)
    axes[countX,countY].set_ylabel('No. of Data Points', fontsize=32, labelpad=6)
    for label in (axes[countX,countY].get_xticklabels() + axes[countX,countY].get_yticklabels()):
        label.set_fontsize(25)
    i += 1
    countY += 1
    if countY == 3:
        countY = 0
        countX += 1

# set the spacing between subplots
plt.tight_layout()
plt.subplots_adjust(wspace=0.24,hspace=0.37)        
plt.show()
```

```python
# Analyzing the skewness in distribution of each modified property
prop1.skew()
```

```python
# Saving the plot of distribution of modified properties
prop1_hist.savefig('Property Histogram-log10.png', dpi=300)
```

```python
# Combining both composition and modified properties dataframe into one dataframe
df = pd.concat([cmp,prop1],axis=1)
df.info()
```

```python
# Calculating the Pearson Correlation coefficients for the combined dataframe
corrmat = df.corr()
corrmat
```

```python
# Plotting the heatmap of the Pearson correlation coefficients of the complete dataframe

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corrmat, dtype=bool))

# Set up the matplotlib figure
heatplot, ax = plt.subplots(figsize=(24, 24))

# Draw the heatmap with the mask and correct aspect ratio
ax = sb.heatmap(corrmat, mask=mask, cmap='vlag_r', vmin=-1, vmax=1, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .7})
ax.tick_params(left=False, bottom=False)

for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(17)
# use matplotlib.colorbar.Colorbar object
cbar = ax.collections[0].colorbar
# here set the labelsize
cbar.ax.tick_params(length=8, labelsize=17, pad = 5)

plt.tight_layout()
plt.show()
```

```python
# Saving the heatmap
heatplot.savefig('Heatmap.png',dpi=300)
```

```python
# Calculating the Pearson correlation coefficients of the modified properties only
prop_corr = prop1.corr()

# Plotting the heatmap of the Pearson correlation coefficients of the modified properties only

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(prop_corr, dtype=bool))

# Set up the matplotlib figure
heatplot_2, ax = plt.subplots(figsize=(20, 20))

# Draw the heatmap with the mask and correct aspect ratio
sb.heatmap(prop_corr, mask=mask, cmap='vlag_r', vmin=-1, vmax=1, center=0, annot = True, fmt = ".2f",
                annot_kws={"size":30}, square=True, linewidths=.5, cbar_kws={"shrink": .7})
ax.set_xticklabels(title)
ax.set_yticklabels(title)
ax.tick_params(left=False, bottom=False)

for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(26)
# use matplotlib.colorbar.Colorbar object
cbar = ax.collections[0].colorbar
# here set the labelsize
cbar.ax.tick_params(length=8, labelsize=22, pad = 8)

plt.tight_layout()
plt.show()
```

```python
# Saving the heatmap of the Pearson correlation coefficients of the modified properties only
heatplot_2.savefig('Prop-prop heatmap.jpg',dpi=300)
```

```python
# Extracting the Pearson correlation coefficients of the modified properties vs elements
c_p_corr = corrmat.iloc[15:,:15]

# Plotting the heatmap of the Pearson correlation coefficients of the modified properties vs elements

# Set up the matplotlib figure
heatplot_3, ax = plt.subplots(figsize=(30, 20))

# Draw the heatmap with the mask and correct aspect ratio
ax = sb.heatmap(c_p_corr, cmap='vlag_r', vmax=1, center=0, annot = True, fmt = ".2f",
                annot_kws={"size":30}, square=True, linewidths=.5, cbar_kws={"shrink": .7})
ax.set_yticklabels(title)
ax.tick_params(left=False, bottom=False)

for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(26)
# use matplotlib.colorbar.Colorbar object
cbar = ax.collections[0].colorbar
# here set the labelsize
cbar.ax.tick_params(length=20, labelsize=22, pad = 8)

plt.tight_layout()
plt.show()
```

```python
# Saving the heatmap of the Pearson correlation coefficients of the modified properties vs elements
heatplot_3.savefig('Comp-prop heatmap.jpg',dpi=300)
```

```python
# Importing the file to cross-check the exported file which will be used for model building and alloy design
full_df = pd.read_excel('FeNiCo_comp-prop.xlsx')
full_df = full_df.iloc[:,1:]
full_df.info()
```

```python
# Removing tensile strength and yield strength columns as Vickers hardness was chosen as the main mechanical property for the study
req_df = pd.concat([full_df.iloc[:,:19], full_df.iloc[:,21:]], axis=1)
req_df
```

```python
# Checking the number of completely filled rows
req_df.dropna()
```

```python
# Counting the number of filled cells in each column
num_df = req_df.count(axis=0)
num_df
```

```python
# Converting the count to percentage
prcnt_df = (num_df/num_df.max()*100).round(2)
prcnt_df
```

```python
# Creating an array of percentage of filled cells in each property column
prcnt_filled = prcnt_df[15:].values
# Appending the percentage value of number of rows/compositions consisting of completely filled property values (8/1208*100 = 0.66)
prcnt_filled = np.append(prcnt_filled,0.66)
prcnt_filled
```

```python
# Creating the labels for the plot of percentage of filled property values
props = [r'$M_s$', r'$log_{10}H_c$', r'$log_{10}T_c$', r'$log_{10}\rho$', r'$log_{10}\delta$', r'$log_{10}H_V$', r'$log_{10}Cost$',r'$Combined$']
props
```

```python
# Plotting the percentage of filled property values in the database
filled, ax = plt.subplots(figsize=(20, 15))

sb.barplot(x=props,y=prcnt_filled)
ax.set_xlabel('Properties', fontsize=28, labelpad=18)
ax.set_ylabel('% filled in database', fontsize =28, labelpad=18)
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(24)
for container in ax.containers:
    ax.bar_label(container, padding = 10, fontsize=22)
ax.tick_params(length=8, pad = 8)

import textwrap
def wrap_labels(ax, width, break_long_words=False):
    labels = []
    for label in ax.get_xticklabels():
        text = label.get_text()
        labels.append(textwrap.fill(text, width=width,
                      break_long_words=break_long_words))
    ax.set_xticklabels(labels, rotation=0)

wrap_labels(ax, 10)

plt.tight_layout()
plt.show()
```

```python
# Saving the plot of percentage of filled property values in the database
filled.savefig('Properties percentage filled in database.jpg',dpi=300)
```

```python
# Creating an array of percentage of empty cells in each property column
prcnt_empty = 100-prcnt_filled
prcnt_empty
```

```python
# Plotting the percentage of empty property values in the database
empty, ax = plt.subplots(figsize=(20, 15))

sb.barplot(x=props,y=prcnt_empty)
ax.set_xlabel('Properties', fontsize=28, labelpad=18)
ax.set_ylabel('% empty in database', fontsize =28, labelpad=18)
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(24)
ax.tick_params(length=8, pad = 8)

import textwrap
def wrap_labels(ax, width, break_long_words=False):
    labels = []
    for label in ax.get_xticklabels():
        text = label.get_text()
        labels.append(textwrap.fill(text, width=width,
                      break_long_words=break_long_words))
    ax.set_xticklabels(labels, rotation=0)

wrap_labels(ax, 10)

plt.tight_layout()
plt.show()
```

```python
# Saving the plot of percentage of empty property values in the database
empty.savefig('Properties percentage empty in database.jpg',dpi=300)
```
